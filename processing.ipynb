```python
import pandas as pd
```

```python
df = pd.read_csv("../breast-cancer-analysis-/breast-cancer.csv",sep=",")
df = df.drop(["id"],axis=1)
```

```python
df['diagnosis'] = (df['diagnosis'] == 'M').astype(int)
```

```python
worst_columns = [item for item in df.columns.to_list() if "worst" in item]

corr = df.corr()

cor_target = abs(corr["diagnosis"])

relevant_features = cor_target[cor_target>0.3]

names = [index for index, value in relevant_features.items()]

final_columns = set(names)-set(worst_columns)
```
